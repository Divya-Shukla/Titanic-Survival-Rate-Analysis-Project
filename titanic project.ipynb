```python
def male_female_kid(traveler):
    # Take the Age and Sex
    age,sex =traveler
    # Compare the age, otherwise leave the sex
    if age < 16:
        return 'kid'
    else:
        return sex
    

# We'll define a new column called 'person', remember to specify axis=1 for columns and not index
titanic_ds['adult or kid'] = titanic_ds[['Age','Sex']].apply(male_female_kid,axis=1)
```
